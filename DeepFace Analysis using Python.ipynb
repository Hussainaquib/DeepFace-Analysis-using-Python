```python
from deepface import DeepFace
import cv2
import matplotlib.pyplot as plt
```

```python
img1=cv2.imread("C:/Users/91975/OneDrive/Pictures/Screenshots/Screenshot 2023-05-30 014221.png")
plt.imshow(img1[:,:,::-1])
plt.show()
```

```python
#Age prediction
result=DeepFace.analyze(img1,actions=['age'])
print("Predicted Age is:", result[0]['age'])
```

```python
#Gender prediction
result = DeepFace.analyze(img1, actions=['gender'])
print("Predicted Gender is:", result[0]['gender'])
```

```python
#Emotion Detection
result = DeepFace.analyze(img1, actions=['emotion'])
print("Predicted Emotion is:", result[0]['dominant_emotion'])
```

```python
#Facial Expression recognition
result = DeepFace.analyze(img1, actions=['emotion'])
print("Predicted Expression is:", result[0]['dominant_emotion'])
```

```python
#Predicted Race/Ethnicity
result = DeepFace.analyze(img1, actions=['race'])
print("Predicted Race/Ethnicity is:", result[0]['dominant_race'])
```
